#### Clean columns with null values

```python
#given the number of columns in the train data, it may not be possible to see each column and the corresponding null value, so a loop was written to only print the coulmns with null value


cols_with_null_values = {}
#loop through each column
for cols in train_dataset.columns:
    
    #count the number of null values
    null_count = train_dataset[cols].isnull().sum()
    
    #a condition to check for columns with null value greater than 0
    if null_count > 0:
        cols_with_null_values[cols] = null_count

#print the columns with null values
print(cols_with_null_values)

print()
#total number of columns with nukk values
print(f'Total number of columns with null value : {len(cols_with_null_values)} ')
```

```python
# Drop  Misc Feature, Feature not aordinal and reature engineer may be challenging 

#train_dataset =train_dataset.drop(columns = ['Misc Feature'])
#train_dataset.head(5)
```

```python
#shape after the feature is dropped
train_dataset.shape
```

```python
# Replace missing categorical values with 'None' to indicate absence of the feature.
# These features are missing not due to data collection issues, but because
# the property simply does not have that characteristic (e.g., no fireplace, no garage, no basement).
# This prevents the model from misinterpreting missing values as meaningful data,
# and helps retain the categorical feature for modeling.

# Specifically:
# Mas Vnr Type: Null indicates no masonry veneer was installed.
# Bsmt Qual / Cond / Exposure / Fin Type: Null indicates the house has no basement.
# Fireplace Qu: Null means no fireplace present.
# Garage Type / Finish / Qual / Cond: Null means no garage exists.

cols_to_fill = ['Mas Vnr Type', 'Bsmt Qual', 'Bsmt Cond', 'Bsmt Exposure', 'BsmtFin Type 1',
                               'BsmtFin Type 2', 'Fireplace Qu', 'Garage Type', 'Garage Finish', 'Garage Qual', 'Garage Cond', 'Alley', 'Pool QC', 'Fence', 'Misc Feature' ]



train_dataset[cols_to_fill] = train_dataset[cols_to_fill].fillna('None')


#confirm if the null value was replaced by None
train_dataset[cols_to_fill].isnull().sum()
```

```python
train_dataset.head(5)
```

```python
# Mas Vnr Area cleaning
# since Mas Vnr Type and Mas Vnr Are are related, confirm if Mas Vnr Area is null where Mas Vnr type is 0. 

Mas_vnr_type = train_dataset[(train_dataset['Mas Vnr Type'] == 'None') & (train_dataset['Mas Vnr Area'].isnull())]
len(Mas_vnr_type)
```

```python
# fill with 0 where Mas Vnr Type	is None
train_dataset.loc[train_dataset['Mas Vnr Type'] == 'None', 'Mas Vnr Area'] = 0

#confirm if null values in Mas Vnr Area has been replaced ny 0

train_dataset['Mas Vnr Area'].isnull().sum()
```

```python
# Fill the columns with 1 0r 2 null valus with 0. This wuld mean the feature is not present,

cols_to_fill = ['BsmtFin SF 1', 'BsmtFin SF 2', 'Bsmt Unf SF', 'Total Bsmt SF', 'Bsmt Full Bath', 'Bsmt Half Bath', 'Garage Cars', 'Garage Area']

train_dataset[cols_to_fill] = train_dataset[cols_to_fill].fillna(0)


#confirm if the null value was replaced by 0
train_dataset[cols_to_fill].isnull().sum()
```

```python
### fill null values in 'Lot Frontage' with median

train_dataset['Lot Frontage'] = train_dataset['Lot Frontage'].fillna(train_dataset['Lot Frontage'].median())
print(train_dataset['Lot Frontage'].isnull().sum())
```

```python
#clean Garage Yr Blt
#since other garage features were replaced with null,  clean Garage Yr Blt will be replaced by 0

train_dataset['Garage Yr Blt'] = train_dataset['Garage Yr Blt'].fillna(0)
print(f" Summ of null value : {train_dataset['Garage Yr Blt'].isnull().sum()}")



# Garage Yr Blt is a float data type. being year, it shouldn't be float
print(f" Garage Yr Blt is a: {train_dataset['Garage Yr Blt'].dtype} data type")

train_dataset['Garage Yr Blt']  = train_dataset['Garage Yr Blt'].astype(int)


print(f" Garage Yr Blt is a: {train_dataset['Garage Yr Blt'].dtype} data type")
```

#### Correlation between numerical features and SalePrice

##### Goal: To identfy the top correlated features

```python
#filter numerical features
numerical_cols = [col for col in train_dataset.columns if train_dataset[col].dtype in ['int64', 'int32', 'float64']]

#calculate correlation between features
correlation_values =  train_dataset[numerical_cols].corr()
```

```python
#get top 10 features most correlated with SalePrice
top_corr = correlation_values['SalePrice'].abs().sort_values(ascending=False)[1:17].index
top_corr_matrix = train_dataset[top_corr.tolist() + ['SalePrice']].corr()
```

```python
# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(top_corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Top Correlated Features with SalePrice')
plt.tight_layout()
plt.show()

# save the graph
plt.savefig("graphs/topcorrelation_heatmap.png")
```

#### Top 15 Correlation Heatmap Interpretation

The above heatmap displays the top 15 numerical features most correlated with the target variable SalePrice (and its log-transformed version Log_SalePrice). 

**Key takeaways include:**

Overall Qual shows the strongest correlation with Log_SalePrice (0.82) and 0.79 with SalePrice, suggesting that the overall material and finish quality of a house is the most powerful predictor of its sale price.
Gr Liv Area (0.71), Total Bsmt SF (0.64), Garage Area (0.64), and Garage Cars (0.66) also exhibit strong linear relationships with price, aligning with expectations that larger and more equipped homes are priced higher.
1st Flr SF, Year Built, and Year Remod/Add fall within a moderate correlation range (0.59–0.61), indicating that recency of construction and renovation have an impact on sale price.
**Garage Area** and **Garage Cars** both have strong positive correlations with Log_SalePrice range (0.64 and 0.66, respectively), but they also appear highly correlated with each other. This would need to be taken into consideration while modeling to avoid multicollinearity. 


```python
# Get 20 least correlated features (excluding SalePrice itself)
bottom_corr = correlation_values['SalePrice'].abs().sort_values(ascending=True)[1:15].index
bottom_corr_matrix = train_dataset[bottom_corr.tolist() + ['SalePrice']].corr()

# Plot heatmap
plt.figure(figsize=(15, 10))
sns.heatmap(bottom_corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Least Correlated Features with SalePrice')
plt.tight_layout()

# Save before showing
plt.savefig("graphs/leastcorrelation_heatmap.png")
plt.show()
```

#### Least 15 Correlation Heatmap Interpretation

The above heatmap presents the features with the weakest correlation to SalePrice, most having coefficients close to zero. These features, such as Overall Cond, Mo old, Yr Sold, Pool Area, BsmtFin SF 2, and Kitchen AbvGr, will be excluded from the model due to their lack of linear relationship with the target variable. Including such variables risks introducing noise or multicollinearity, which may reduce model performance or interpretability. 
Feature selection focused on keeping predictors with clear, strong, and justifiable influence on house price.

Instead, better-performing features with clearer linear trends and stronger correlations were prioritized. Pool Area for example has a highly right-skewed with most values at zero, indicating that the vast majority of homes do not have a pool.

### Feature Engineering

#### Target variable transformation

```python
##perform a log transformation to normalize the ditsribution of the SalePrice

train_dataset['Log_SalePrice'] = np.log(train_dataset['SalePrice'])

plt.figure(figsize=(8, 4))
plt.hist(train_dataset['Log_SalePrice'], bins=40, edgecolor='black')
plt.title('Log-Transformed SalePrice Distribution')
plt.xlabel('Log(SalePrice)')
plt.ylabel('Count')
plt.tight_layout()
plt.show()

#save the figure
plt.savefig('graphs/Log_saleprice.png')
```

The log-transformed version of `SalePrice` appears much more **symmetrical and normally distributed** compared to the original distribution. This transformation effectively reduces right skewness and compresses extreme values, making the data more suitable for linear regression modeling.

### Log transformation of numerical variables. 'Lot Area', 'Mas Vnr Area', 'BsmtFin SF 1', 'Garage Area' etc.
'Open Porch SF', 'Wood Deck SF', 'Enclosed Porch', 'Screen Porch',
'Pool Area', 'Low Qual Fin SF', '2nd Flr SF', 'Lot Frontage' is observed to be right skwed wih concentration of values near 0. Usig np.log(df[col]),the feature will be more evenly distributed. 

```python
# Log-transform skewed numerical features
skewed_features = ['Gr Liv Area', 'Lot Area', 'Mas Vnr Area', 'BsmtFin SF 1', 'Garage Area',
    'Open Porch SF', 'Wood Deck SF', 'Enclosed Porch', 
    'Screen Porch', 'Pool Area', 'Low Qual Fin SF', 'Lot Frontage']

# Create a copy of the numerical features (to avoid modifying the original)
numerical_log = train_dataset[skewed_features].copy()
for col in skewed_features:
    numerical_log[col] = np.log1p(numerical_log[col])
```

#### Categorical variales transformation

The features below will be selected for feature transformation based on potential impact on SalePrice. Including all categorical feature may introduce noise, hence, the reason for removing the other features. Misc Feature was added as some unque homes may standout, there by impacting the price of the house.

**'MS Zoning', 'Neighborhood', 'Condition 1', 'Condition 2', 'Bldg Type',
                          'House Style', 'Roof Style', 'Roof Matl', 'Exterior 1st',
                          'Exterior 2nd', 'Mas Vnr Type', 'Foundation', 'Heating',
                          'Central Air', 'Garage Type', 'Sale Type', 'Street', 'Alley', 'Misc Feature',
                                        'Land Contour'**



 



```python
# List of categorical columns to encode
categorical_cols = [
    'MS Zoning', 'Neighborhood', 'Condition 1', 'Condition 2', 'Bldg Type',
    'House Style', 'Roof Style', 'Roof Matl', 'Exterior 1st',
    'Exterior 2nd', 'Mas Vnr Type', 'Foundation', 'Heating',
    'Central Air', 'Garage Type', 'Sale Type', 'Street', 'Alley',
    'Misc Feature', 'Land Contour'
]

# Step 1: Select ONLY categorical columns
categorical_data = train_dataset[categorical_cols]

# Step 2: One-hot encode JUST those columns
categorical_encoded = pd.get_dummies(
    categorical_data,
    columns=categorical_cols,
    drop_first=True,
    dtype=int
)

# Verify (should now ONLY contain one-hot encoded columns)
print("Shape:", categorical_encoded.shape)
categorical_encoded.head()
```

```python
X = pd.concat([
    categorical_encoded,      # One-hot encoded features
    numerical_log,            # Log-transformed numericals
    train_dataset[label_encoded_cols]], axis=1) #label encoded features

#log of Saleprice
y = train_dataset['Log_SalePrice']
```

#### RMSE and R2

The model was evaluated using Root Mean Squared Error (RMSE) and R-squared (R²), with the target variable SalePrice log-transformed to reduce skewness and stabilize variance.

The RMSE is 0.145, indicating that the average prediction error is approximately 0.145 in the log scale. The R² score is 0.865, meaning the model explains about 86.5% of the variance in the log-transformed SalePrice. Since the target variable is log-transformed, the RMSE reflects error in log-dollars. To interpret this in actual dollar terms, the error can be transformed by exponentiating the RMSE: i.e exp(0.145) = 1.156

```python
# Lasso Regression

lasso_params = {'alpha': [0.0001, 0.001, 0.01, 0.1, 1, 10]}
lasso = Lasso(max_iter=10000)
lasso_grid = GridSearchCV(lasso, lasso_params, cv=5, scoring='neg_root_mean_squared_error')

lasso_grid.fit(X_train_scaled, y_train)

print("Best Lasso alpha:", lasso_grid.best_params_)
print("Best Lasso RMSE:", -lasso_grid.best_score_)

# perform Prediction on test set
y_pred_lasso = lasso_grid.predict(X_test_scaled)

rmse_lasso = np.sqrt(mean_squared_error(y_test, y_pred_lasso))
r2_lasso = r2_score(y_test, y_pred_lasso)

print(f"Lasso Regression : RMSE: {rmse_lasso:.3f}, R²: {r2_lasso:.3f}")
```

### Lasso Regression Evaluation

Lasso regression was used to evaluate model performance with L1 regularization, which can shrink less important feature coefficients to zero and thus help with feature selection.

Using GridSearchCV, the best alpha value was found to be 0.001. With this regularization strength, the model achieved the following results on the validation set:

Root Mean Squared Error (RMSE): 0.135

R-squared (R2): 0.883

This indicates that the model explains approximately 88.3% of the variance in the log-transformed SalePrice, and the average prediction error is relatively low. The low RMSE and high R² suggest that Lasso performs well in capturing the relationship between predictors and the target while potentially improving generalizability by reducing overfitting through coefficient shrinkage.



```python
#Ridge regression

ridge_params = {'alpha': [0.1, 1, 10, 100, 500, 1000]}
ridge = Ridge()
ridge_grid = GridSearchCV(ridge, ridge_params, cv=5, scoring='neg_root_mean_squared_error')
ridge_grid.fit(X_train_scaled, y_train)

print("Best Ridge alpha:", ridge_grid.best_params_)
print("Best Ridge RMSE:", -ridge_grid.best_score_)

# Predict on test set
y_pred_ridge = ridge_grid.predict(X_test_scaled)

rmse_ridge = np.sqrt(mean_squared_error(y_test, y_pred_ridge))
r2_ridge = r2_score(y_test, y_pred_ridge)

print(f"Ridge Regression : RMSE: {rmse_ridge:.3f}, R²: {r2_ridge:.3f}")
```

Ridge regression was applied using L2 regularization, which helps reduce model complexity by shrinking coefficients without eliminating them entirely.

Using GridSearchCV, the optimal alpha value was determined to be 100. At this level of regularization, the model achieved the following performance on the validation set:

Root Mean Squared Error (RMSE): 0.136

R-squared (R²): 0.881

This means the model explains approximately 88.1% of the variance in the log-transformed SalePrice, with a low average prediction error.
